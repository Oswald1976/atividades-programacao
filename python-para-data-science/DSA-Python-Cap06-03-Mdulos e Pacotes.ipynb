## Pacotes e Módulos

Em Python, um m´dulo é um arquivo(script) que contém código Python e pode ser importado em outros arquivos Python. Ele é usado para compartilhar funçõesc classes e váriáveis entre arquivos.
Já um pacote é uma coleção de módulos organizados em uma estrutura de diretórios. Ele permite a divisão de um aplicativo em múltiplos módulos, o que facilita a manutenção e o desenvolvimento.
Visite o Pypi, repositório de Pacotes da Linguagem Python: https:/pypi.org/

Podemos trabalhar com módulos dos pacotes (quando disponíveis)

```python
# importando o módulo request do pacote urllib, usado para trazer url´s
# para dentro do nosso ambiente Python
import urllib.request
```

```python
# Variavel resposta armazena o objeto da conexão a url passada como parâmetro
resposta = urllib.request.urlopen('https://portaldoservidor.guarulhos.sp.gov.br/emp_wa.php#')
```

```python
print(resposta) # Objeto retornado do comando anterior (ponteiro onde fica esta resposta)
```

```python
# chama o método read() do objeto retornado e armazena na variavél todo o código html
html = resposta.read()
```

```python
# imprime o código html armazenado na variavél que utilizou o metodo read()
print(html)
```
