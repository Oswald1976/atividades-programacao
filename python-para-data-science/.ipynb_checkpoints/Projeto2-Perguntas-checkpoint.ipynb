## Pergunta de Negócio 1:

### Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

```python
# Primeiro Filtramos o dataframe com os registros da categoria que desejamos
df_dsa_pl = df_dsa[df_dsa['Categoria'] == 'Office Supplies']
```

```python
# Em seguida agrupamos por cidade e calculamos o total de Valor_Venda
df_dsa_p1_total = df_dsa_pl.groupby('Cidade')['Valor_Venda'].sum()
```

```python
# Então encontramos a cidade com maior valor de venda
cidade_maior_venda = df_dsa_p1_total.idxmax()
print('Cidade com maior valor de venda para Office Supplies :', cidade_maior_venda)
```

```python
# Para conferir o resultado
df_dsa_p1_total.sort_values(ascending = False)
```

## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
# Agrupamos por segmento e calculamos o total de vendas
df_dsa_p5 = df_dsa.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_values(by = 'Valor_Venda', ascending = False)
```

```python
df_dsa_p5.head()
```

```python
# Função para coverter os dados em valor absoluto
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return ' $ {v:d}'.format(v = val)
    return my_format
        
```

```python
# Plot

# Tamanho da figura
plt.figure(figsize = (16, 6))

# Grafico de Pizza
plt.pie(df_dsa_p5['Valor_Venda'],
        labels = df_dsa_p5['Segmento'],
        autopct = autopct_format(df_dsa_p5['Valor_Venda']),
        startangle = 90)

# Limpa o Circulo Central
centre_circle = plt.Circle((0, 0), 0.82, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Labels e anotações
plt.annotate(text = 'Total de Vendas: '+ ' $ ' + str(int(sum(df_dsa_p5['Valor_Venda']))), xy = (-.35, 0))
plt.title('Total de Vendas por Segmento')
plt.show()
```

## Pergunta de Negócio 8 (Desafio Nível Master):

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
# Criar uma nova coluna de acordo com a regra definida acima
df_dsa['Desconto'] = np.where(df_dsa['Valor_Venda'] > 1000, 0.15, 0.10)
```

```python
df_dsa.head(13)
```

```python
# Criamos uma coluna calculando o valor de venda menos o desconto
df_dsa['Valor_V_Desconto'] = df_dsa['Valor_Venda'] - (df_dsa['Valor_Venda'] * df_dsa['Desconto'])
```

```python
df_dsa.head()
```

```python
# Filtrando as vendas antes do desconto de 15%
df_dsa_p8_ven_ant_desc = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda']
```

```python
# Filtrando as vendas depois do desconto de 15%
df_dsa_p8_ven_dep_desc = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_V_Desconto']
```

```python
# Calcula a media das vendas antes do desconto de 15%
media_ven_ant_desc = df_dsa_p8_ven_ant_desc.mean()
```

```python
# Calcula a media das vendas depois do desconto de 15%
media_ven_dep_desc = df_dsa_p8_ven_dep_desc.mean()
```
